# clean-papers-bibtex

This script cleans a bibtex generated by Papers3. In particular it removes lines that reference to Papers3 or its files.

```python
BIBTEX_FILE = "/<path-to-file>"
```

```python
with open(BIBTEX_FILE, "r") as bibtex_in:
    lines = bibtex_in.readlines()

new_lines = [line for line in lines
             if not (line.startswith("uri = {\\url{papers3") or 
                     line.startswith("local-url = ") or 
                     line.startswith("file = "))]

with open(BIBTEX_FILE + '.cleaned', 'w') as bibtex_out:
    bibtex_out.writelines(new_lines)
```
